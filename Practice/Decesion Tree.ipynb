```python
# standardization
X=df_ohe.drop('Attrition_Flag',axis=1)
y=df_ohe['Attrition_Flag']
df_ohe_std=(X-X.mean())/X.std()
```

```python
df_ohe_std.describe()
```

```python
# outliers detection

def impute_outlier(df,col):
    df.loc[df[col]>3,col]=3
    df.loc[df[col]<-3,col]=-3
    return df

df_impute=df_ohe_std
for col in df_impute:
    df_impute=impute_outlier(df_impute,col)


df_impute
```
