# Encoding
#### Converting string data to integer or float type data as Most of machine learning algo read this type of data
- ordinal data (Label Encoding)
- nominal data (onehot encoding)

```python
df.head()
```

```python
df.info()
```

```python
df['BusinessTravel']
```

```python
df['BusinessTravel'].unique()
```

```python
label_map={'BusinessTravel':{'Travel_Rarely':0,'Travel_Frequently':1,'Non-Travel':2}}
df.replace(label_map).head()
```

# onehot Encoding

```python
df_ohe=pd.get_dummies(df,columns=['Attrition','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'])
```
